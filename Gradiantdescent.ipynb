```python
class SGD:
    def __init__(self):  # <-- fixed double underscores
        self.m = -120
        self.b = 100

    def fit(self, xtrain, ytrain, lr, epochs):
        for i in range(epochs):
            loss_slope_b = -2 * np.sum(ytrain - self.m * xtrain.ravel() - self.b)
            self.b = self.b - loss_slope_b * lr
            loss_slope_m = -2 * np.sum((ytrain - self.m * xtrain.ravel() - self.b)*xtrain.ravel())
            self.m = self.m - loss_slope_m * lr
    def predict(self,xtest):
        ypred=self.m*xtest+self.b
        return ypred
```

```python
import numpy as np
from sklearn.metrics import r2_score
from sklearn.preprocessing import StandardScaler

class SGD:
    def __init__(self, n_features):
        # initialize weight vector and bias
        self.w = np.zeros(n_features)
        self.b = 0.0

    def fit(self, X, y, lr=0.01, epochs=1000):
        n = X.shape[0]
        for i in range(epochs):
            # predicted y
            y_pred = np.dot(X, self.w) + self.b

            # gradients
            dw = (-2/n) * np.dot(X.T, (y - y_pred))
            db = (-2/n) * np.sum(y - y_pred)

            # update parameters
            self.w -= lr * dw
            self.b -= lr * db

            # optional: monitor training progress
            if i % 100 == 0:
                loss = np.mean((y - y_pred) ** 2)
                print(f"Epoch {i}: Loss = {loss:.4f}")

    def predict(self, X):
        return np.dot(X, self.w) + self.b
```

```python
# assume Xtrain, Xtest, ytrain, ytest are defined

# feature scaling (recommended for SGD)
scaler = StandardScaler()
xtrain = scaler.fit_transform(xtrain)
xtest = scaler.transform(xtest)

# initialize model with number of features
gd = SGD(n_features=xtrain.shape[1])

# train
gd.fit(xtrain, ytrain, lr=0.01, epochs=1000)

# predict
ypred = gd.predict(xtest)

# evaluate
print("Weights (w):", gd.w)
print("Bias (b):", gd.b)
print("R2 Score:", r2_score(ytest, ypred))
```
