```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample dataset
df = pd.DataFrame({
    "Age": [22, 25, 47, 52, 46, 56],
    "Salary": [20000, 32000, 50000, 60000, 52000, 58000]
})

# Scatter plot (Age vs Salary)
plt.scatter(df["Age"], df["Salary"], color='blue', marker='o')
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
plt.hist(df["Salary"], bins=10, color='orange')
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Count")
plt.show()
```

```python
models = ['KNN', 'SVM', 'RandomForest', 'XGBoost']
accuracy = [0.85, 0.88, 0.91, 0.93]
plt.bar(models, accuracy,color=['red','green','yellow','orange'])
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")
plt.show()
```

```python
plt.bar(models, accuracy, color=['red', 'green', 'blue', 'purple'])
plt.title("Model Accuracy Comparison")
plt.ylabel("Accuracy")

for i, v in enumerate(accuracy):
    plt.text(i, v + 0.005, f"{v:.2f}", ha='center')

plt.show()
```
